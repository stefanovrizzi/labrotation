```python
plt.figure
trial = mat["EMG_array"][0][0][0][0][0][0]
plt.plot(range(0,np.size(trial)),trial)
plt.title('Trial figure')
plt.show()
```

Squeeze all traces

```python
for j in range(0,np.size(mat["EMG_array"][0][0][0][0][0])):
    for i in range(0,np.size(mat["EMG_array"][0][0][0][0][0])):
        mat["EMG_array"][0][0][0][0][j][i] = np.squeeze(mat["EMG_array"][0][0][0][0][j][i])
```

Plot multiple-trace trial figure

```python
plt.figure
for i in range(0,np.size(mat["EMG_array"][0][0][0][0][0])):
    trial = mat["EMG_array"][0][0][0][0][0][i]
    plt.plot(range(0,np.size(trial)),trial)
plt.title('Trial figure 2')
plt.show()
```

Limit boundaries to zoom-in on signal

```python
plt.figure
for i in range(0,np.size(mat["EMG_array"][0][0][0][0][0])):
    trial = mat["EMG_array"][0][0][0][0][0][i]
    plt.plot(range(0,np.size(trial)),trial, alpha=.2)
plt.title('Trial figure 2 - zoom-in')
plt.ylim([-2000, 2000])
plt.show()
```
